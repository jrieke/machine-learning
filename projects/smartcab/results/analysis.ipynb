```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
import pandas as pd
```

```python
results = pd.read_csv('coarse_mean_results.csv')
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scat = ax.scatter(results['epsilon'], results['alpha'], results['gamma'], s=200*results['success']**2, c=results['success'], vmin=0, vmax=1)
ax.set_xlabel('epsilon')
ax.set_ylabel('alpha')
ax.set_zlabel('gamma')
fig.colorbar(scat, label='Success rate', pad=0.15)

plt.show()

# plt.savefig('coarse.png', dpi=300)
```

```python
plt.plot(results.groupby('epsilon').max()['success'], 'o-', label='epsilon')
plt.plot(results.groupby('alpha').max()['success'], 'o-', label='alpha')
plt.plot(results.groupby('gamma').max()['success'], 'o-', label='gamma')
plt.ylim(0.5, 1)
plt.ylabel('max success rate')
plt.legend()
```
